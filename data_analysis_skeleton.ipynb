```python
# Run gpt_survey_analysis to rate all survey responses for your survey and store ratings in approrpriate files.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
import scipy.stats as stats
import textwrap
```

```python
def violinPlot(sequence1, sequence2, pre_text, post_text):
    data = sequence1 + sequence2
    # Create a 'category' list to differentiate between sequences
    categories = ['Pre-course survey'] * len(sequence1) + ['Post-course survey'] * len(sequence2)

    # Colors
    colors = ['skyblue', 'lightcoral']

    # Violin plot
    plt.figure(figsize=(12,7))
    sns.violinplot(x=categories, y=data, inner=None, palette=colors, alpha=0.8)

    sns.stripplot(x=categories, y=data, jitter=True, marker="o", edgecolor="black", linewidth=0.5, size=10, palette=colors, order=['Pre-course survey', 'Post-course survey'])

    wrapped_pre = "Pre-course: "+textwrap.fill(pre_text, 120)
    wrapped_post = "Post-course: " + textwrap.fill(post_text, 120)

    wrapped_title = '\n'.join([wrapped_pre, wrapped_post])

    plt.title(wrapped_title, fontsize=15, loc='left')
    plt.ylabel('Value', fontsize=13)
    plt.xlabel('')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    p_val_a = stats.ttest_ind(sequence1, sequence2, nan_policy='omit')[1]
    plt.text(1.35, max(data) * 0.95, f'p = {p_val_a:.1e}', horizontalalignment='center', verticalalignment='center', fontsize=18)
```

```python
pre_post_survey = "/path/to/data.xlsx"
pre_post_survey_data = pd.read_excel(pre_post_survey, header=0)

pre_post_survey_dict = {}
for column in pre_post_survey_data.columns:
    pre_post_survey_dict[column] = pre_post_survey_data[column].tolist()

print(pre_post_survey_dict.keys())

```
